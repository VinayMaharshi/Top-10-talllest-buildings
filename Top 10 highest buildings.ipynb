```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
df=pd.read_csv(r"D:\Data Science\data sets\tallest_buildings_global.csv")
df.head(10)
```

```python
df.isnull().head(10)
```

```python
plt.figure(figsize=(12,8)) #declaring the figure size.


def addlabels(x,y): #creating a function to call labels on bars.
    for i in range(len(x)):
        plt.text(i,y[i],y[i],fontname='Times New Roman'
                 ,weight='bold',fontsize=14,ha='center',va='bottom')
 

#declaring the values we want to plot as x and y.
x=df['name'].head(10) 
y=df['height_m'].head(10)

#plotting a bargraph with the values of x and y.
plt.bar(x,y)

#setting title and label name to the graph.
plt.title('Top 10 tallest buildings across the world in meters.'
          ,fontname='Times new roman',fontsize=20,weight='bold',pad=20)

addlabels(x,y) #calling the funstion.

plt.xlabel('Name of the Building' ,fontname='Times new roman',fontsize=20
           ,weight='bold',labelpad=20,color='red')
plt.ylabel('Height in Meters' ,fontname='Times new roman',fontsize=20
           ,weight='bold',labelpad=20,color='red')

#setting style to the graph ticks.
plt.xticks(rotation=90,fontname='Times new roman',fontsize=14,weight='bold')
plt.yticks(fontname='Times new roman',fontsize=14,weight='bold')

plt.show()
```
